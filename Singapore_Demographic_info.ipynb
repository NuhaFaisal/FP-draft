## Data Preprocessing

```python
import pandas as pd

df = pd.read_csv("/content/IndicatorsOnPopulation.csv")
```

```python
df.keys()
```

```python
df['Data']
```

```python
df['Data'] = pd.to_datetime(df['Data'], format='%Y')
df['Data']
```

```python
df.set_index('Data', inplace=True)
```

```python
df.index
```

```python
df.info()
```

```python
print(df.columns)
```

```python
import numpy as np
```

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].replace('na', np.nan)
    df[col] = df[col].fillna('-1')
    df[col] = df[col].astype(float)
    df[col] = df[col].replace(-1, np.nan)

    print(f"Column: {col}")
    print(df[col].dtype)
    print(df[col].unique())
```

```python
df.info() # all cols covert into numrical
```

```python
df.isna().sum()
```

```python
df
```

```python
df.dropna(inplace=True)
```

## regions

```python
df_region = pd.read_csv("/content/region.csv")
```

```python
df_region
```

```python
df_region.index.max()
```

```python
df_region['Data'] = pd.to_datetime(df_region['Data'], format='%Y')
df_region['Data']
```

```python
df_region.set_index('Data', inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
for column in df_region.columns:
    plt.plot(df_region.index, df_region[column], label=column)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population by Region in Singapore 2019-2023')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_2023 = df_region.loc['2023-01-01']

plt.figure(figsize=(10, 8))
plt.pie(df_2023, labels=df_2023.index, autopct='%1.1f%%', colors=plt.cm.Paired(range(len(df_2023))))

plt.title('Population Distribution by Region in 2023')
plt.show()
```

```python
df_Transport = pd.read_csv('/content/Transport_to_work.csv')
df_Transport
```

```python
df_pie = df_Transport[df_Transport['Transport_to_Work'] != 'Total']

plt.figure(figsize=(10, 8))
plt.pie(df_pie['Total'], labels=df_pie['Transport_to_Work'], autopct='%1.1f%%', colors=plt.cm.Paired(range(len(df_pie))))

plt.title('Population Distribution by Transport Mode')
plt.show()
```

```python
df_top5 = df_pie.nlargest(5, 'Total')

plt.figure(figsize=(10, 8))
plt.pie(df_top5['Total'], labels=df_top5['Transport_to_Work'], autopct='%1.1f%%', colors=plt.cm.Paired(range(len(df_top5))))


plt.title('Top 5 Transport Modes by Population in 2023')
plt.show()
```
